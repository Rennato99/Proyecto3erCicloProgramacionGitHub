```python
# Standard imports 
from bokeh.io import output_notebook
from bokeh.plotting import figure, output_file, output_notebook, show
import pandas as pd
import numpy as np
output_notebook()
```

```python
# Abrimos el archivo csv y lo convertimos en un DataFrame
dataFrame01 = pd.read_csv("data/EstimatesOfUnderFiveMortality.csv", sep=";", engine='python')
```

```python
dataFrame01 # Imprimos el dataFrame
```

```python
# Necesitamos solo el promedio de muertes por cada region
averageDict = {}
averageDict['Region Name'] = []
i = 3
while(i<=47):
    averageDict['Region Name'].append(dataFrame01['Region Name'][i])
    i += 4


```

```python
year = 1990.5

while(year <= 2016.5):
    averageDict[str(year)] = []
    
    i = 3
    while(i<=47):
        
        averageDict[str(year)].append(dataFrame01[str(year)][i])
        i += 4
    
    year += 1

```

```python
print(averageDict) 
```

```python
# Convertimos el diccionario de los promedio en un nuevo DataFrame
dataFrame = pd.DataFrame(averageDict)
```

```python
dataFrame # Imprimimos el nuevo dataFrame
```

```python
# Calculamos un total de muertes por año y lo agregamos como una columna al final de dataFrame


generalAverage = {'Averages': []}
# Recorremos cada fila
index = 0
while index <= 11:
    year = 1990.5
    
    suma = 0 
    # Recorremos cada columna
    while(year <= 2016.5):
        suma += dataFrame[str(year)][index]      
        year += 1
    average = np.round((suma / 28), 2)
    generalAverage['Averages'].append(average)
    
    index += 1
# Obtenemos un diccionario:  {'Averages': [130.34, 147.66, 113.48, 38.53, 82.63, 33.75, 30.82, 7.89, 18.84, 30.08, 5.92, 66.1]}

# Lo agregamos como columna a nuestro dataFrame
dataFrame['Averages'] = generalAverage['Averages']
```

```python
# Imprimimos el dataFrame para comprobar como ha quedado
dataFrame
```

```python
"""
Procedemos a crear las figuras con bokeh
"""
```

```python

x_range = (dataFrame['Region Name'])
print(x_range)

figure01 = figure(x_range=x_range, plot_height=250, title="Under Five Mortality", toolbar_location=None, tools="")

figure01.vbar(x=x_range, top=dataFrame['Averages'], width=0.9)
figure01.y_range.start = 0

show(figure01)
```
