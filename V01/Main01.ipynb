```python
# Standard imports 
from bokeh.io import output_notebook
from bokeh.plotting import figure, output_file, output_notebook, show
import pandas as pd
import numpy as np
output_notebook()


from bokeh.models import ColumnDataSource, HoverTool
from bokeh.sampledata.autompg import autompg_clean as df
from bokeh.transform import factor_cmap

# Abrimos el archivo csv y lo convertimos en un DataFrame
dataFrame01 = pd.read_csv("data/Estimates-of-under-five mortality.csv", sep=";", engine='python')

# Declaramos una paleta de colores
colors01 = ['#EF7A63', '#BFF14A', '#5EB3BB', '#21738C', '#792E9C', '#8C8EC9', '#17A655', '#C4C81C', '#BDBE90', '#DAA37B', 
            '#D9371B', '#5C81EA', '#04F7DA', '#833235', '#437032', '#FFD500', '#B2FF00', '#FF9C00', '#FFD8EE', '#E4D8FF', 
            '#D8DFFF', '#D8FAFF', '#D8FFE2', '#F3FFD8', '#FFE7D8', '#6B6464', '#813C3C']

# Calculamos un total de muertes por año y lo agregamos como una columna al final de dataFrame


generalAverage = {'Averages': []}
# Recorremos cada fila
index = 0
while index <= 35:
    year = 1990.5
    
    sum_ = 0 
    # Recorremos cada columna
    while(year <= 2016.5):
        sum_ += dataFrame01[str(year)][index]      
        year += 1
    average = np.round((sum_ / 28), 2)
    generalAverage['Averages'].append(average)
    
    index += 1
# Obtenemos un diccionario:  {'Averages': [130.34, 147.66, 113.48, 38.53, 82.63, 33.75, 30.82, 7.89, 18.84, 30.08, 5.92, 66.1]}

# Lo agregamos como columna a nuestro dataFrame
dataFrame01['Averages'] = generalAverage['Averages']


# Agrupamos segun el nombre de la region
group = dataFrame01.groupby('Region Name', as_index = False).mean()
group
```

```python
"""
Procedemos a crear las figuras con bokeh
"""
```

```python
# la primera figura sera barra


x_range = group['Region Name']
y_range = group['Averages']


figure01 = figure(x_range=x_range, plot_height=550, title="Under Five Mortality", toolbar_location=None, tools="")
#  figure01.add_tools(HoverTool(tooltips=[("Region Name:", "@Region Name"), ("Averages", "@Averages")]))
figure01.add_tools(HoverTool(tooltips=[("Region Name:", "Region"), ("Averages", "@Averages")]))
figure01.vbar(x=x_range, top=y_range, width=0.9, line_color=colors01[2], line_width=2, fill_color=colors01[:12])
figure01.y_range.start = 0
figure01.yaxis.axis_label = "Number of children"
figure01.xaxis.axis_label = "Region"
figure01.xaxis.major_label_orientation = 1.2
output_file(filename='Figure01.html', title='Figure01', mode='cdn', root_dir=None)
show(figure01)
```

```python
# Ahora crearemos un grafico de pastel usando wedge
import math
pi = math.pi
```

```python
# Metodo para obtener el porcentaje de un numero
def getPercentage(total, part):
    return (part*100)/total

def getAngle(percentage):
    return (2*pi*percentage)/(100)
```

```python
# Lo metemos dentro de un ciclo
index = 0

# Creamos la figura
figure05 = figure(plot_width=1000, plot_height=1000)    
group03 = group.drop(group.index[11])

total = group03.sum()['Averages']# Variable que servira para obtener el porcentaje

start_angle = 0 # El valor inicial del angula será 0
end_angle = 0  # El valor inicial del angula será 0
while index <= 10:
    
    
    # Obtenemos el promedio
    part = group03.iloc[index]['Averages']
 
    # Obtenemos el porcentaje de su promedio
    percentage = getPercentage(total, part)

    # Obtenemos el valor en radianes
    end_angle += getAngle(percentage)

    # Leyenda
    legend = ("%s - %.2f " %(str(group03['Region Name'][index]), percentage))
    legend += " %"

    # Agregamos un wedge
    figure05.wedge(x=[500], y=[500], radius=300, start_angle=start_angle,
           end_angle=end_angle, radius_units="screen", color=colors01[index], legend=legend, direction="anticlock")
    
    
    
    
    
    # El angulo final de este wedge será el inicio del siguiente
    start_angle=end_angle

    # Aumentamos el contador
    index+=1

group03
show(figure05)
```
